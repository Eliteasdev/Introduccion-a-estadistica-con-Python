## <span style="color:#50fa7b;">Ejercicio 1</span>
Comprobar gráficamente si las ecuaciones tien solución.
* Sistema 1
<br>$3x+2y=6$
<br>$6x+4y=-9$
* Sistema 2
<br>$x+y=3$
<br>$x+4y=1$
* Sistema 3
<br>$2x-y+2z=3$
<br>$2x+3y-z=0$

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from mpl_toolkits import mplot3d

#Primer sistema de ecuaciones
x = np.linspace(-2.0, 2.0)
fig,ax = plt.subplots()
l1 = ax.plot(x, (6 - (3*x)) / 2, color="#ff79c6", label="3x+2y=6")
l2 = ax.plot(x, (-9 - (6*x)) / 4, color="#50fa7b", label="6x+4y=-9")
plt.legend()
plt.grid()
plt.show()
```

```python
#Segundo sistema de ecuaciones
x = np.linspace(0, 6.0)
fig,ax = plt.subplots()
l1 = ax.plot(x, 3 - x, color="#ff79c6", label="x+y=3") 
l2 = ax.plot(x, (1 - x) / -4, color="#50fa7b", label="x-4=1")
plt.legend()
plt.grid()
plt.show()
```

## <span style="color:#50fa7b;">Ejercicio 2</span>
Encuentra la ecuación de la recta tangente a la parabola $y=x²+3$ en el punto $P(1,4)$.

```python
x = np.linspace(-10,10,400) #Genera los valores en el rango necesario
x_p = 1
y_p = 4


y = x**2 + 3
dy_dx = np.gradient(y, x) # Derivada con respecto a x
m = dy_dx[np.abs(x - x_p).argmin()] # Obtener el valor de la pendiente en el punto (1,4)
b = y_p - m * x_p # Utilizar la ecuacion de la recta de la forma y=mx+b para encontrar el valor de b

fig,ax = plt.subplots()
l1 = ax.plot(x, x**2 + 3, color="#ff79c6", label="Parabola x**2 + 3") 
l2 = ax.plot(x, m * x + b, color="#50fa7b", label="Tangente en P(1,4)")
p1 = ax.scatter(x_p, y_p, color="red", label="Punto(1,4)")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Recta tangente a la parabola")
plt.grid(True)
plt.show()
```
